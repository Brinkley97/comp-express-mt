```python
one_to_many_df = df.iloc[:, [0, 1]]
many_to_one_df = df.iloc[:, [3, 4]]
```

```python
import pandas as pd

source = []
last_valid_value = None

for idx, row in one_to_many_df.iterrows():
    source_row = row['English (Source, One)']
    print(f"{idx}: {source_row}")
    
    if pd.isna(source_row):
        print("\tMissing value detected.")
        if last_valid_value is not None:
            source_row = last_valid_value
    else:
        last_valid_value = source_row  # Update the last known good value
    
    source.append(source_row)
```

```python
source
```

```python
one_to_many_df["English (Source, One)'"] = source
one_to_many_updated_df = one_to_many_df.iloc[:, [1, 2]]
one_to_many_updated_df
```

```python
one_to_many_pt = pd.pivot_table(one_to_many_updated_df, index=["English (Source, One)'", "Akan (Target, Many)"], aggfunc='first')
one_to_many_pt
```
